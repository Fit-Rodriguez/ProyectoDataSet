<h5 style =" color:#00b990">4.3 Imputacion de valores numericos</h5>

```python
numeric_cols = df2.select_dtypes(include=[np.number])
mean_values = numeric_cols.mean()
df_filled = df2.fillna(mean_values)

print("DataFrame con valores faltantes imputados:")
print(df_filled)
```

```python
numeric_cols = df2.select_dtypes(include=[np.number])
df2.head(2)
```

```python
av_serialNumber = df2["serialNumber"].astype("float").mean(axis=0)
av_listYear = df2["listYear"].astype("float").mean(axis=0)



```

```python
df2["serialNumber"].replace(np.nan,av_serialNumber,inplace = True)
df2["listYear"].replace(np.nan,av_listYear,inplace = True)
df2["dateRecorded"].replace(np.nan,av_listYear,inplace = True)
df2["assessedValue"].replace(np.nan,av_listYear,inplace = True)
df2["salesRatio"].replace(np.nan,av_listYear,inplace = True)



df2.head(5)
```

```python
df2["assessedValue"] = pd.to_numeric(df2["assessedValue"],errors = "coerce")
df2["saleAmount"] = pd.to_numeric(df2["saleAmount"],errors = "coerce")
df2["salesRatio"] = pd.to_numeric(df2["salesRatio"],errors = "coerce")


```

```python
df2.info()
```

```python
numeric_columns = df2.select_dtypes(include=['float64', 'int64']).columns

av_means = {}
for column in numeric_columns:
    av_means[column] = df2[column].mean()

print(av_means)
```

<h5 style =" color:#00b990">4.4 Imputacion de valores categoricos</h5>


```python
df2["listYear"].value_counts().idxmax()
```

```python
df2["listYear"].replace(np.nan,'', inplace = True)
df2.head(5)
```

```python
df = pd.DataFrame(df2)
# Identificar columnas categóricas
categorical_cols = df2.select_dtypes(include=['object'])

print("Columnas categóricas:")
print(categorical_cols)

# Imputación de valores categóricos (usando 'Desconocido')
for column in categorical_cols:
    df[column] = df[column].fillna('Desconocido')

print("\nDataFrame con valores categóricos imputados:")
print(df)





```
